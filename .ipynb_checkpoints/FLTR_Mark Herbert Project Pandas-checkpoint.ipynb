```python
###########ENTAIN##############
#I will now read in a further dataset and compare this (Entain and Flutter) 

# Read the CSV file
df_ENT = pd.read_csv('ENT.L.csv', parse_dates=['Date'],  dayfirst=True,index_col='Date')
```

```python
#summary statistics for the stocks 
summary = pd.DataFrame({
    'Flutter_mean': df_FLTR.mean(),
    'Entain_mean': df_ENT.mean(),
    'Flutter_std': df_FLTR.std(),
    'Entain_std': df_ENT.std()})
summary
```

```python
pd.set_option('display.max_rows', 10) and pd.set_option('display.max_rows', 10) #settting the total number of rows. 

df_ENT = df_ENT.round(2) #rounds to 2 decimal places.
```

```python
plt.plot(df_FLTR.index, df_FLTR['Close'], label='Flutter') #this shows us general trends but is not ideal as the two stocks start at different places
plt.plot(df_ENT.index, df_ENT['Close'], label='Entain')
plt.legend()
plt.show()
```

```python
#Adding in the same columns as are in df_FLTR 

df_ENT['Month'] = df_ENT.index.month

df_ENT['Close_by_Month'] = df_ENT['Close'].groupby(df_ENT['Month']).transform('mean')

df_ENT['Trend'] = np.where(df_ENT['Close'] > df_ENT['Close'].shift(1), 'Up', 'Down')

print(df_ENT.head())
```

```python
# I will try to normalise the data by dividing each value by the first value in the series, this should ensure that both stocks start at the same place
df_FLTR_norm = df_FLTR['Close'] / df_FLTR['Close'].iloc[0]
df_ENT_norm = df_ENT['Close'] / df_ENT['Close'].iloc[0]
```

```python
# Line plot of the 'Closed' price comparison of both stocks with the normalised data
plt.plot(df_FLTR_norm, label='Flutter Stock', linestyle='--')
plt.plot(df_ENT_norm, label='Entain Stock')
plt.xlabel('Time')
plt.ylabel('Normalized Stock Price')
plt.title('Line Plot of Normalized Stock Prices')
plt.legend()
plt.show()
```

```python
df_FLTR_norm_2022 = df_FLTR_norm[df_FLTR_norm.index.year == 2022]
df_ENT_norm_2022 = df_ENT_norm[df_ENT_norm.index.year == 2022]

plt.plot(df_FLTR_norm_2022, label='Flutter Stock', linestyle='--')
plt.plot(df_ENT_norm_2022, label='Entain Stock')
plt.xlabel('Time')
plt.ylabel('Normalized Stock Price')
plt.title('Line Plot of Normalized Stock Prices for 2022')
plt.legend()
plt.show()
```

```python
df_FLTR_norm_2018 = df_FLTR_norm[df_FLTR_norm.index.year == 2018]
df_ENT_norm_2018 = df_ENT_norm[df_ENT_norm.index.year == 2018]

plt.plot(df_FLTR_norm_2018, label='Flutter Stock', linestyle='--')
plt.plot(df_ENT_norm_2018, label='Entain Stock')
plt.xlabel('Time')
plt.ylabel('Normalized Stock Price')
plt.title('Line Plot of Normalized Stock Prices for 2018')
plt.legend()
plt.show()
```

```python
#Also using the normalised data I have the two stocks 'Close' distributuion on a Histogram.

# Plot histogram of FLTR
plt.hist(df_FLTR_norm, bins=30, alpha=0.5, label='Flutter')

# Plot histogram of ENT 
plt.hist(df_ENT_norm, bins=30, alpha=0.4, label='Entain')

#legend added
plt.legend()

# Display plot
plt.show()
```

```python
# Scatter plot comparison with normalised data
plt.scatter(df_FLTR_norm, df_ENT_norm)
plt.xlabel('Flutter Price')
plt.ylabel('Entain Price')
plt.title('Scatter Plot of Flutter & Entain Stock Prices')
plt.show()
```

```python
plt.fill_between(df_FLTR.index, df_FLTR_norm, df_ENT_norm, where=df_FLTR_norm > df_ENT_norm, interpolate=True, alpha=0.5)
plt.fill_between(df_FLTR.index, df_ENT_norm, df_FLTR_norm, where=df_FLTR_norm < df_ENT_norm, interpolate=True, alpha=0.5)
plt.xlabel('Time')
plt.ylabel('Normalized Stock Price')
plt.title('Area Plot of Normalized Stock Prices')
plt.show()
```

```python
#Plotting the volume of Flutter and Entain on two lines
plt.plot(df_FLTR.index, df_FLTR['Volume'], color='red')
plt.plot(df_FLTR.index, df_ENT['Volume'], color='blue')

#  labels
plt.xlabel('Days')
plt.ylabel('Volume, #')
plt.title('Volume Comparison between Flutter & Entain')
plt.show()
```

```python
# I will try to normalize the volume now like I did with the close price to be able to compare a lot clearer
df_FLTR_normvol = df_FLTR['Volume'] / df_FLTR['Volume'].iloc[0]
df_ENT_normvol = df_ENT['Volume'] / df_ENT['Volume'].iloc[0]
```

```python
#Looking at corellation between the two stocks, for this I will merge the two datasets
#Merge the dataframes on 'Date' and used the stock names as suffixes.

merged_df = pd.merge(df_ENT, df_FLTR, left_index=True, right_index=True, suffixes=('_ENT', '_FLTR'))
```

```python
print(merged_df.head())
```

```python
print(merged_df.info())
```

```python
# Calculate the correlation between the 'adjusted close' columns of each stock
corr = merged_df['Close_ENT'].corr(merged_df['Close_FLTR'])
corr1 = merged_df['Open_ENT'].corr(merged_df['Open_FLTR'])
corr2 = merged_df['High_ENT'].corr(merged_df['High_FLTR'])
corr3 = merged_df['Low_ENT'].corr(merged_df['Low_FLTR'])
corr4 = merged_df['Volume_ENT'].corr(merged_df['Volume_FLTR'])

print("The 'Close' price correlation between ENT and FLTR is", corr)
print("The 'Open' price correlation between ENT and FLTR is", corr1)
print("The 'High' price correlation between ENT and FLTR is", corr2)
print("The 'Low' price correlation between ENT and FLTR is", corr3)
print("The 'Volume' price correlation between ENT and FLTR is", corr3)
```

```python
#This now focuses in closer with the pl.xlim() function

df_FLTR_normvol = df_FLTR['Volume'] / df_FLTR['Volume'].iloc[0]
df_ENT_normvol = df_ENT['Volume'] / df_ENT['Volume'].iloc[0]

sns.histplot(df_FLTR_normvol, label='Flutter Stock', kde=True, binwidth=0.05, alpha=0.5)
sns.histplot(df_ENT_normvol, label='Entain Stock', kde=True,  binwidth=0.05, alpha=0.5)

plt.legend()
plt.title('Comparisons of Normalised Volume Distribution between Flutter and Entain Stock')
plt.xlabel('Normalized Volume')
plt.xlim(0.2, 10)
plt.show()
```

```python
df_FLTR_normvol = df_FLTR['Volume'] / df_FLTR['Volume'].iloc[0]
df_FLTR_normvol = df_FLTR_normvol.reset_index()
df_FLTR_normvol['Year'] = df_FLTR_normvol['Date'].apply(lambda x: x.year)
df_FLTR_normvol['Norm_Volume'] = df_FLTR_normvol['Volume']
sns.histplot(data=df_FLTR_normvol, x='Norm_Volume', hue='Year', multiple="stack")
plt.xlabel('Normalized Volume')
plt.ylabel('Count')
plt.xlim(0.2, 10)
plt.title('Normalized Volume Distribution by Year')
plt.show()
```

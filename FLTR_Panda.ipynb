```python
# Calculate the price change
delta = df_FLTR['Close'].diff()

# Define the time window for RSI calculation
window_length = 40

# Calculate the gains and losses with exponential weighting
gain = delta.where(delta > 0, 0)
loss = -delta.where(delta < 0, 0)
avg_gain = gain.ewm(com=(window_length - 1), min_periods=window_length).mean()
avg_loss = loss.ewm(com=(window_length - 1), min_periods=window_length).mean()

# Calculate the relative strength
rs = avg_gain / avg_loss

# Calculate the RSI
rsi = 100 - (100 / (1 + rs))

# Add the RSI to the DataFrame
df_FLTR['rsi'] = rsi

# Visualize the RSI with matplotlib
df_FLTR['rsi'].plot()
plt.show()
```
